```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "unsupervised_learning"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

Read data from the “hw07_data.csv”. This is a multiple feature dataset.

```python
X = np.loadtxt("hw07_data.csv", delimiter=",")
```

Run the Scikit-Learn standard KMeans algorithm on the dataset. Set n_clusters to
40 and output the inertia value.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=40)
kmeans.fit(X)
print(kmeans.inertia_)
```

Write a loop and run standard KMeans algorithm with different cluster
numbers on the data. The cluster number should range from 5 to 150 with step of 5. 

```python
inertia_values = []
models = []
k_values = []
i = 5
while i <= 150:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    k_values.append(i)
    models.append(kmeans)
    inertia_values.append(kmeans.inertia_)
    i = i + 5
```

Collect the inertia values and visualize the results with K values. (inertia
values vs K values) 

```python
# The elbow method
plt.figure()
plt.plot(k_values, inertia_values, "bo-")
plt.xlabel("$K$", fontsize=14)
plt.ylabel("Inertia", fontsize=14)         
save_fig("inertia_vs_K_plot")
plt.show()
```

Could you find the elbow point? What do you think the best K value is?<br>
I do not believe I can find the elbow point in this graph. At k=40 the decrease in inertia per additional 5 clusters becomes noticeably smaller and incremental from there on, however teh score can still be reduced by more than half so there may be a more valid cluster value further down. I would say a k value of 35 or 40 would be reasonable based off of this graph and that only 280 rows exist with the dataset.

Collect the silhouette scores of the above models and visualize the results
with K values.

```python
from sklearn.metrics import silhouette_score
silhouette_scores = [silhouette_score(X, model.labels_)
                     for model in models]
plt.figure()
plt.plot(k_values, silhouette_scores, "bo-")
plt.xlabel("$K$", fontsize=14)
plt.ylabel("Silhouette Score", fontsize=14)
save_fig("silhouette_score_vs_K_plot")
plt.show()
```

Could you find the best K value using this method? What do you think the
best K value is? <br>
Out of all the K values, it appears that the best point is at k=90. There are no higher points from there onward. I do have the concern that there are 280 points of data and 90 clusters, and the Silhoutte Score is still relatively low at k=90. k=40 was picked in the previous plot for the elbow method, and does show a significant increase in Silhouette Score compared to lower K values. From there the return per increment is lower but still can increase by nearly 50% by teh time k=90 is achieved. 

 Load the picture “sea.jpg” to the project and display it.

```python
from matplotlib.image import imread
image = imread("sea.jpeg")
plt.imshow(image)
plt.axis('off')
plt.show()
```

Process the data, so you can apply KMeans algorithm to cluster the data.

```python
image = image/255.0 # convert to 0...1 scale
X = image.reshape(-1, 3)
X.shape
```

Run Mini-Batch KMeans algorithm on the data with cluster number K = 16
and K = 8.

```python
from sklearn.cluster import MiniBatchKMeans
minibatch8 = MiniBatchKMeans(n_clusters = 8)
minibatch8.fit(X)
minibatch_colors8 = minibatch8.cluster_centers_[minibatch8.predict(X)]
minibatch_colors8 = minibatch_colors8.reshape(image.shape)

minibatch16 = MiniBatchKMeans(n_clusters = 16)
minibatch16.fit(X)
minibatch_colors16 = minibatch16.cluster_centers_[minibatch16.predict(X)]
minibatch_colors16 = minibatch_colors16.reshape(image.shape)
```

Display the original picture and processed pictures. The three pictures
should be displayed in the same row with proper labels.

```python
plt.figure(figsize=(16,9))
plt.subplots_adjust(wspace=0.05)
plt.subplot(131)
plt.imshow(image)
plt.title("Original image")
plt.axis('off')

plt.subplot(132)
plt.imshow(minibatch_colors8)
plt.title("8-color image")
plt.axis('off')

plt.subplot(133)
plt.imshow(minibatch_colors16)
plt.title("16-color image")
plt.axis('off')

save_fig('image_segmentation_diagram', tight_layout=False)
plt.show()
```
