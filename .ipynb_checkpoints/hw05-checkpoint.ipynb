First, let's import a few common modules, ensure MatplotLib plots figures inline and prepare a function to save the figures. We also check that Python 3.5 or later is installed (although Python 2.x may work, it is deprecated so we strongly recommend you use Python 3 instead), as well as Scikit-Learn ≥0.20. You don't need to change any code in the following cell, just run it.

```python
# Python ≥3.5 is required
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os

# to make this notebook's output stable across runs
np.random.seed(42)

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)

# Where to save the figures
PROJECT_ROOT_DIR = "."
CHAPTER_ID = "decision_trees"
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images", CHAPTER_ID)
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

 Read data from the “hw05_data.csv”. 

```python
# Import pandas to read in the csv
import pandas as pd

# Save the data  
data = pd.read_csv("hw05_data.csv")
```

Train a decision tree model based on the data:

```python
# Encode the data with one hot encoding
data_encoded = pd.get_dummies(data, drop_first=True)

# Split features and outcome
X = data_encoded.iloc[:, 0:3]
y = data_encoded.iloc[:, 3]

# Import sklearn 
from sklearn.tree import DecisionTreeClassifier

# Create the Tree
tree_clf = DecisionTreeClassifier(random_state=42)

# Fit the Data
tree_clf.fit(X, y)
```

Display the Decision Tree

```python
# Import the necessary libraries
from graphviz import Source
from sklearn.tree import export_graphviz

# Call Export Graphvis for classifier
export_graphviz(
        tree_clf,
        out_file=os.path.join(IMAGES_PATH, "exercise_tree.dot"),
        feature_names=X.columns,
        class_names=["Yes", "No"],
        rounded=True,
        filled=True
    )

Source.from_file(os.path.join(IMAGES_PATH, "exercise_tree.dot"))
```

 Display the prediction result with the following input:

```python
# Call the prediction for Sunny, Just Ate, and Worked Late
pred = tree_clf.predict([[1,1,1]])[0]

# Conditionally Assign a readable output
status = "Yes" if pred == 1 else "No"

# Print the Prediction
print("The predicted output for exercising is: %s"%(status))
```
